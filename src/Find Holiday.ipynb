```python
from datetime import datetime, timedelta
import pandas as pd
import os
```

```python
def Todate(date_str) : 
    date_object = datetime.strptime(date_str, '%Y-%m-%d').date()
 #   print(type(date_object))
 #   print(date_object) 
    return date_object
```

```python
Holiday2016=["2016-01-01","2016-02-07","2016-02-08","2016-02-09","2016-02-10","2016-04-13"
             ,"2016-05-05","2016-05-06","2016-05-14","2016-06-06","2016-06-13","2016-08-15",
             "2016-09-14","2016-09-15","2016-09-16","2016-10-03","2016-10-09","2016-12-25"]
Holiday2017=["2017-01-01","2017-01-27","2017-01-28","2017-01-30",
             "2017-05-03","2017-05-05","2017-06-06","2017-06-13","2017-08-15",
             "2017-10-03","2017-10-04","2017-10-05","2017-10-06","2017-10-09","2017-12-25"]
Holiday2018=["2018-01-01","2018-02-15","2018-02-16","2018-02-17","2018-05-05","2018-05-07",
            "2018-05-22","2018-06-06","2018-06-13","2018-08-16","2018-09-23","2018-09-24","2018-09-25","2018-09-26",
            "2018-10-03","2018-10-09","2018-12-25"]
Holiday_data=list()
Holiday_data.extend(Holiday2016)
Holiday_data.extend(Holiday2017)
Holiday_data.extend(Holiday2018)
count =0
for time in makerange:
    for hol in Holiday_data:
        datehol=Todate(hol)
        if datehol == time:
            count+=1
print(count)
```
