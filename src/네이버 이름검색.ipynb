```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import requests
import json
import pandas as pd
from tqdm.notebook import tqdm
```

```python
data = pd.read_csv("data/movie_df.csv")  #2012/01/01~2020/02/09년도 박스오피스 데이터
data.head()
data.tail()
```

```python
colums_to_save=["movieCd","movieNm"]
total =data[colums_to_save].drop_duplicates()
print(total["movieCd"].shape,"\n",total.head(10))
```

```python
lpoint=[]
driver=webdriver.Chrome("C:/ChromeDriver/chromedriver.exe")

for name in tqdm(total["movieNm"]):
    movieNm=name
    driver.get("https://movie.naver.com/movie/search/result.nhn?query="+movieNm+"&section=movie&ie=utf8")
    try:
        point=driver.find_element_by_xpath("//*[@id='old_content']/ul[2]/li[1]/dl/dd[1]/em[1]")
    except:
        print("점수를 찾을수 없는 영화입니다. : ",name)
        
        lpoint.append(0)
        continue
    else:
        lpoint.append(point.text)
        #print(point.text)
        #point=''
        #rint(name," : ",point.text)
driver.close()
total["point"]=lpoint
total.head(10)

```

```python
total.to_csv("data/박스오피스와평균평점.csv")
total.head(20)
```
